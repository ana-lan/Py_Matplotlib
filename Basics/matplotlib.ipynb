##### Adding fill to line plots

````python
#load data
data = pd.read_csv('data1.csv')
ages = data['Age']
dev_sal = data['All_Devs']
py_sal = data['Python']
js_sal = data['JavaScript']
````

````python
#plotting All_Devs and Py
plt.style.use('fivethirtyeight')
plt.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal, label='Python')

plt.legend()
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
````

````python
#Adding fill
plt.style.use('fivethirtyeight')
plt.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal, label='Python')

#used for the fill
#alpha for transparency
plt.fill_between(ages, py_sal,alpha = 0.25)

plt.legend()
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
````

````python
#fill between a value and line
overall_median = 57287

plt.style.use('fivethirtyeight')
plt.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal, label='Python')

#adding the y values for the fill
plt.fill_between(ages, py_sal, overall_median, alpha = 0.25)

plt.legend()
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
````

````python
#conditional filling
plt.style.use('fivethirtyeight')
plt.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal, label='Python')

#adding the y values for the fill
#interpolate to ensure no x-intersections are left out and all sections are filled correctly
plt.fill_between(ages, py_sal, overall_median, 
                    where = (py_sal > overall_median),
                    interpolate = True, alpha = 0.25)

plt.legend()
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.tight_layout()
````

````python
#conditional different color filling
plt.style.use('fivethirtyeight')
plt.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_sal, label='Python')

#adding the y values for the fill
#interpolate to ensure no x-intersections are left out and all sections are filled correctly
plt.fill_between(ages, py_sal, overall_median, 
                    where = (py_sal > overall_median),
                    interpolate = True, alpha = 0.25)

plt.fill_between(ages, py_sal, overall_median, 
                    where = (py_sal <= overall_median),
                    interpolate = True, alpha = 0.25)

plt.legend()
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.tight_layout()
````

#### Read data from csv file and plot

````python
#read and manipulate data according to our need
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    row = next(csv_reader)
    #getting the desired data
    #all the languages sperated by names
    print(row['LanguagesWorkedWith'].split(';'))
````

````python
#using the counter method to calculate the data for languages
#wil count the nummber of repititions of a word in the data

#defining an empty counter:
lang_counter = Counter()

#read and manipulate data according to our need
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    #using the counter on every row of the csv file to count the number of occurences of a language
    for row in csv_reader:
        lang_counter.update(row['LanguagesWorkedWith'].split(';'))

print(lang_counter)
````

````python
#using the counter method: most_common(<value>) to get just the most common 15 values in the counter
print(lang_counter.most_common(15))
#returns tuples
````

````python
#get the data for x and y axes in a list
lang = []
popular = []

for item in lang_counter.most_common(15):
    lang.append(item[0])
    popular.append(item[1])
print(lang)
print(popular)
````

````python
#plotting the graph
plt.bar(lang, popular)

plt.title("Most Popular programming languages")
plt.xlabel("Languages")
plt.ylabel("Numbe rrof users")

plt.tight_layout
plt.show()
````

````python
#creating a horizontal bar chart
plt.barh(lang, popular)

plt.title("Most Popular programming languages")
plt.ylabel("Languages")
plt.xlabel("Number rof users")

plt.tight_layout
plt.show()
````

````python
#print the bar chart in reverse order:

#reversing the lists
lang.reverse()
popular.reverse()

plt.barh(lang, popular)

plt.title("Most Popular programming languages")
plt.ylabel("Languages")
plt.xlabel("Number rof users")

plt.tight_layout
plt.show()
````

````python
#using pandas to deal with the csv file management:
data = pd.read_csv('data.csv')
ids = data['Responder_id']
responses = data['LanguagesWorkedWith']

language_counter = Counter()
for response in responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

#repeated code: 
for item in lang_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

#plot
languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most Popular programming languages")
plt.ylabel("Languages")
plt.xlabel("Number rof users")

plt.tight_layout
plt.show()
````

### Real Time Data Plots

````python
#data
x_vals = []
y_vals = []

#creating realtime data
#counter variable
index = count()

#function to create values
def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))

    #clears the previous data befor plotting new line
    plt.cla()
    plt.plot(x_vals, y_vals)


#using the FuncAnimation function of matplotlib for the data of every second
#passing the current figure to it
#using the animate opertaion on the figure
#for every 1 sec: interval = 1000 (since argument is of 1 millisecond)
ani = FuncAnimation(plt.gcf(), animate, interval = 1000)

plt.tight_layout()
plt.show()
````

plotting on multiple axes

````python
#to share the xticks the sharex is used
fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols = 1, sharex=True)

#plotting on multiple axes
ax1.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, py_sal, label='Python')
ax2.plot(ages, js_sal, label='JavaScript')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
````

plotting 2 diff figs

````python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

#plotting on multiple figs
ax1.plot(ages, dev_sal, color='#444444',
         linestyle='--', label='All Devs')
ax1.legend()
ax2.set_xlabel('Ages')
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.plot(ages, py_sal, label='Python')
ax2.plot(ages, js_sal, label='JavaScript')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

#to save those figures
fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
````
